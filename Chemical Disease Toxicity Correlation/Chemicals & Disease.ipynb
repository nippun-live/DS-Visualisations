```python
df2 = df[df['_unit_state'] == "finalized"].reset_index()
df2
```

```python
df2 = df2.drop(columns=['_golden', '_unit_state', '_trusted_judgments','_last_judgment_at', 'level_0', 'index', 'original_job_id', 'verify_relationship_gold', 'verify_relationship_gold_reason', 'orig__golden'])
df2
```

```python
df3 = df2[(df2["verify_relationship"] =="yes_direct")|(df2["verify_relationship"] == "yes_indirect")].reset_index()
df3 = df3.drop(columns=['index'])
df3
```

```python
df_agg = df3.groupby(by=['chemical_name', 'disease_name']).agg("mean").reset_index()
df_agg
```

```python
df_agg.nlargest(10, 'verify_relationship:confidence')
```

```python
df_test = df3.groupby(by=['chemical_name', 'disease_name']).agg("sum").reset_index()
df_test.nlargest(10, 'verify_relationship:confidence')
```
